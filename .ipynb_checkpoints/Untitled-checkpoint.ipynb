```python
class node:

    def __init__(self):
        #instance fields found by C++ to Python Converter:
        self.arr = [[0 for _ in range(8)] for _ in range(8)]
        self.row = 0
        self.column = 0
        self.next = None

def main():
    a = None
    a = node()
    for i in range(0, 8):
        for j in range(0, 8):
            a.arr[i][j] = 0
    a.row = -1
    a.next = None
    Globals.Queen8(a)

class Globals:

    @staticmethod
    def print(a):
        for i in range(0, 8):
            for j in range(0, 8):
                print(a.arr[i][j], end = '')
                print(" ", end = '')
            print("\n", end = '')
        print("\n", end = '')
    @staticmethod
    def valid(a):
        #vartical
        i = 0
        while i<(a.row):
            if a.arr[i][a.column] == 1:
                #cout<<"\nfalse 1\n"
                return False
            i += 1
        #digonal1
        i = a.row-1
        j = a.column-1
        while i>=0 and j>=0:
            if a.arr[i][j] == 1:
                #cout<<"\nfalse 2\n"
                return False
            i -= 1
            j -= 1
        #diagonal2
        i = a.row-1
        j = a.column+1
        while i>=0 and j>=0:
            if a.arr[i][j] == 1:
                #cout<<"\nfalse 1\n"
                return False
            i -= 1
            j += 1
        #cout<<"\nTRUE\n"
        return True
    @staticmethod
    def Queen8(a):
        last = None
        last = node()
        t = 1
        while a.row<8:
            for i in range(0, 8):
                b = None
                b = node()
                for j in range(0, 8):
                    for k in range(0, 8):
                        b.arr[j][k] = a.arr[j][k]
                b.row = a.row+1
                b.column = i
                b.arr[b.row][i] = 1
                if t == 1 and Globals.valid(b):
                    a.next = b
                    last = b
                    last.next = None
                    t = 0
                elif Globals.valid(b):
                    last.next = b
                    last = b
                    last.next = None
                #print(b)
            if a.next is not None:
                a = a.next
            #print(a)
            if a.row == 7:
                Globals.print(a)
                return
            #cout<<"\n"<<a->row

if __name__ == "__main__":
    main()
```
